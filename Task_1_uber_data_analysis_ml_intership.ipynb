```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

#step2 loading datasets
url = "https://raw.githubusercontent.com/Geo-y20/Uber-Rides-Data-Analysis/main/UberDataset.csv"
dataset = pd.read_csv(url)

dataset.head()
```

```python
dataset.shape
```

```python
dataset.info()
```

```python
# data preprocessing and EDA(Exploratory data analysis)
```

```python
dataset['PURPOSE'].fillna('Not',inplace=True)
```

```python
dataset
```

```python
dataset['START_DATE']=pd.to_datetime(dataset['START_DATE'],errors='coerce')
```

```python
dataset['END_DATE']=pd.to_datetime(dataset['END_DATE'],errors='coerce')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
from datetime import datetime
```

```python
dataset['Date']=pd.DatetimeIndex(dataset['START_DATE']).date
dataset.head()
```

```python
dataset['Time']=pd.DatetimeIndex(dataset['START_DATE']).hour
dataset.head()
```

```python
dataset['Day_Night']=pd.cut(x=dataset['Time'],bins=[0,10,15,19,24],labels=['Morning','Afternoon','Evening','Night'])
```

```python
dataset.head()
```

```python
dataset.dropna(inplace=True)
dataset.shape
```

```python
# 1 graph visualization
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)

sns.countplot(dataset['CATEGORY'])
plt.xticks(rotation=90)

# 2
plt.subplot(1,2,2)
sns.countplot(dataset['PURPOSE'])
plt.xticks(rotation=90)
```

```python
sns.countplot(dataset['Day_Night'])
```

```python
dataset.head()
```

```python
# 4
dataset['Month']=pd.DatetimeIndex(dataset['START_DATE']).month
```

```python
month_label={
    1.0:'jan',2.0:'feb',3.0:'march',4.0:'April',5.0 :'may',
    6.0:'june',7.0:'july',8.0:'aug',9.0:'sep',10.0:'oct',
    11.0:'nov',12.0:'dec'
}
dataset['Month']=dataset['Month'].map(month_label)
```

```python
mon=dataset.Month.value_counts(sort=False)
dataset.head()
```

```python
df=pd.DataFrame({
    "month":mon.values,
    "value_count":dataset.groupby('Month',sort=False)['MILES'].max
})
```

```python

p=sns.lineplot(data=df)
p.set(xlabel='months',ylabel='value_counts')
```

```python
dataset.head()
```

```python
dataset['Day']=dataset.START_DATE.dt.weekday

day_labels={
    0.0:'mon',1.0:'Tue',2.0:'wed',3.0:'thu',4.0:'fri',5.0:'sat',6.0:'sun'
}
dataset['Day']=dataset['Day'].map(day_labels)
```

```python
day_labels=dataset.Day.value_counts()
```

```python
sns.barplot(x=day_labels.index, y=day_labels)
plt.xlabel('Day')
plt.ylabel('Count')
```
